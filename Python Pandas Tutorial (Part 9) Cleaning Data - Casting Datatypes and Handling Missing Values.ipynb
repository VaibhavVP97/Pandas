```python
import pandas as pd
import numpy as np
```

```python
na_vals = ['NA', 'Missing']
df = pd.read_csv('Data.csv', index_col='Respondent', na_values=na_vals)
schema_df = pd.read_csv('survey_results_schema.csv', index_col='Column')
```

```python
people = {
    'first': ['Corey', 'Jane', 'John', 'Chris', np.nan, None, 'NA'], 
    'last': ['Schafer', 'Doe', 'Doe', 'Schafer', np.nan, np.nan, 'Missing'], 
    'email': ['CoreyMSchafer@gmail.com', 'JaneDoe@email.com', 'JohnDoe@email.com', None, np.nan, 'Anonymous@email.com', 'NA'],
    'age': ['33', '55', '63', '36', None, None, 'Missing']
}
```

```python
dff = pd.DataFrame(people)
dff
```

```python
dff.dropna()
```

```python
dff.dropna(axis='index',how='any')
```

```python
dff.dropna(axis='index',how='all')
```

```python
dff
```

```python
dff.dropna(axis='columns',how='any')
```

```python
dff.dropna(axis='columns',how='all')
```

```python
dff.dropna(axis='index',how='all',subset=['email'])
```

```python
dff.dropna(axis='index',how='any',subset=['email'])
```

```python
dff.dropna(axis='index',how='all',subset=['last','email'])
```

```python
dff.replace('NA',np.nan,inplace=True)
dff.replace('Missing',np.nan,inplace=True)

dff
```

```python
dff.isna()
```

```python
dff.fillna(0)
```
